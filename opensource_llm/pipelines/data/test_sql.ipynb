```python
### load data from northwind.db to sqlite3. show the tables and their schema
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('northwind.db')

# Create a cursor object
cursor = conn.cursor()

# Query to get the list of tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

# Print the list of tables
print("Tables in the database:")
for table in tables:
    print(table[0])

# Show the schema of each table
for table in tables:
    print(f"\nSchema of table {table[0]}:")
    cursor.execute(f"PRAGMA table_info({table[0]});")
    schema = cursor.fetchall()
    for column in schema:
        print(column)

# Close the connection
conn.close()
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('northwind.db')
cursor = conn.cursor()

# Function to rename tables and columns to lower case
def rename_to_lowercase(cursor):
    # Get the list of tables
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    for table in tables:
        table_name = table[0]
        new_table_name = table_name.lower()

        # Rename the table if the new name is different
        if table_name != new_table_name:
            cursor.execute(f"ALTER TABLE {table_name} RENAME TO {new_table_name};")

        # Get the schema of the table
        cursor.execute(f"PRAGMA table_info({new_table_name});")
        schema = cursor.fetchall()

        for column in schema:
            column_name = column[1]
            new_column_name = column_name.lower()

            # Rename the column if the new name is different
            if column_name != new_column_name:
                cursor.execute(f"ALTER TABLE {new_table_name} RENAME COLUMN {column_name} TO {new_column_name};")

# Rename tables and columns to lower case
rename_to_lowercase(cursor)

# Commit the changes and close the connection
conn.commit()
conn.close()
```
